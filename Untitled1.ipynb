```python
import numpy as np

print("=" * 60)
print("üîß FEATURE ENGINEERING")
print("=" * 60)

# –í—Ä–µ–º–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
df['hour'] = df['step'] % 24
df['is_night'] = ((df['hour'] >= 0) & (df['hour'] <= 5)).astype(int)
df['is_weekend'] = ((df['step'] // 24) % 7 >= 5).astype(int)

# –ë–∞–ª–∞–Ω—Å–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
df['balance_change_orig'] = df['newbalanceOrig'] - df['oldbalanceOrg']
df['amount_to_balance'] = df['amount'] / df['oldbalanceOrg'].replace(0, 1)
df['amount_to_balance'] = np.clip(df['amount_to_balance'], 0, 10)

print(f"‚úÖ –°–æ–∑–¥–∞–Ω–æ {df.shape[1]-10} –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
```

```python
print("=" * 60)
print("üö® –°–û–ó–î–ê–ù–ò–ï –ü–†–ê–í–ò–õ –ü–û –¢–ó")
print("=" * 60)

# 1. –ù—É–ª–µ–≤–æ–π –±–∞–ª–∞–Ω—Å + –∫—Ä—É–ø–Ω–∞—è —Å—É–º–º–∞ (>1 –º–ª–Ω)
df['rule1'] = ((df['oldbalanceOrg'] == 0) & (df['amount'] > 1000000)).astype(int)

# 2. –°—É–º–º–∞ > 2 –º–ª–Ω
df['rule2'] = (df['amount'] > 2000000).astype(int)

# 3. –í—ã–≤–æ–¥ –≤—Å–µ—Ö –¥–µ–Ω–µ–≥ + —Å—É–º–º–∞ >500K
df['rule3'] = ((df['amount_to_balance'] == 10) & (df['amount'] > 500000)).astype(int)

# 4. –ù–æ—á—å + –∫—Ä—É–ø–Ω–∞—è —Å—É–º–º–∞ (>1 –º–ª–Ω)
df['rule4'] = ((df['is_night'] == 1) & (df['amount'] > 1000000)).astype(int)

# 5. –û—á–µ–Ω—å –±—ã—Å—Ç—Ä—ã–µ –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–∏ (<1 —á–∞—Å–∞)
df_sorted = df.sort_values(['nameOrig', 'step'])
df_sorted['prev_step'] = df_sorted.groupby('nameOrig')['step'].shift(1)
df_sorted['time_since_last'] = df_sorted['step'] - df_sorted['prev_step']
df = df_sorted.copy()
df['rule5'] = (df['time_since_last'] < 1).astype(int)

# 6. –ü–æ–ª—É—á–∏–ª —Å—Ä–∞–∑—É –æ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –æ—Ç–ø—Ä–∞–≤–∏—Ç–µ–ª–µ–π (—Å–ª–æ–∂–Ω–æ –±–µ–∑ –ø–æ–ª–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö, —É–ø—Ä–æ—â–∞–µ–º)
# –î–ª—è –ø–æ–ª—É—á–∞—Ç–µ–ª—è —Å—á–∏—Ç–∞–µ–º —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–∏ –≤ –ø–æ—Å–ª–µ–¥–Ω–∏–π —á–∞—Å
df['rule6'] = 0  # –ó–∞–≥–ª—É—à–∫–∞

# 7. –û—Ç–ø—Ä–∞–≤–∏–ª —Å—Ä–∞–∑—É –Ω–µ—Å–∫–æ–ª—å–∫–∏–º –ø–æ–ª—É—á–∞—Ç–µ–ª—è–º
df['rule7'] = 0  # –ó–∞–≥–ª—É—à–∫–∞

# –û–±—â–∏–π —Å—á–µ—Ç –ø—Ä–∞–≤–∏–ª
df['rule_score'] = df['rule1'] + df['rule2'] + df['rule3'] + df['rule4'] + df['rule5']
df['rule_based_suspicious'] = (df['rule_score'] >= 2).astype(int)

print(f"üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø—Ä–∞–≤–∏–ª:")
print(f"   ‚Ä¢ –í—Å–µ–≥–æ —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–π: {len(df):,}")
print(f"   ‚Ä¢ –ü–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö –ø–æ –ø—Ä–∞–≤–∏–ª–∞–º: {df['rule_based_suspicious'].sum():,}")
print(f"   ‚Ä¢ –ü—Ä–∞–≤–∏–ª–æ 1 (–Ω—É–ª–µ–≤–æ–π –±–∞–ª–∞–Ω—Å + >1M): {df['rule1'].sum():,}")
print(f"   ‚Ä¢ –ü—Ä–∞–≤–∏–ª–æ 2 (—Å—É–º–º–∞ >2M): {df['rule2'].sum():,}")
print(f"   ‚Ä¢ –ü—Ä–∞–≤–∏–ª–æ 3 (–≤—Å–µ –¥–µ–Ω—å–≥–∏ + >500K): {df['rule3'].sum():,}")
print(f"   ‚Ä¢ –ü—Ä–∞–≤–∏–ª–æ 4 (–Ω–æ—á—å + >1M): {df['rule4'].sum():,}")
print(f"   ‚Ä¢ –ü—Ä–∞–≤–∏–ª–æ 5 (–±—ã—Å—Ç—Ä—ã–µ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏—è): {df['rule5'].sum():,}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score
import time

print("=" * 60)
print("ü§ñ ML –ú–û–î–ï–õ–¨ - RANDOM FOREST")
print("=" * 60)

start = time.time()

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è ML
features = [
    'amount', 'oldbalanceOrg', 'newbalanceOrig',
    'oldbalanceDest', 'newbalanceDest', 'amount_to_balance',
    'balance_change_orig', 'hour', 'rule_score'
]

X = df[features].fillna(0)
y = df['isFraud']

# –°–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –≤—ã–±–æ—Ä–∫–∞
fraud_data = df[df['isFraud'] == 1]
normal_data = df[df['isFraud'] == 0].sample(n=len(fraud_data)*10, random_state=42)
train_df = pd.concat([fraud_data, normal_data])

X_train = train_df[features].fillna(0)
y_train = train_df['isFraud']

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    class_weight={0: 1, 1: 10}
)
model.fit(X_train, y_train)

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è –≤—Å–µ—Ö –¥–∞–Ω–Ω—ã—Ö
df['ml_probability'] = model.predict_proba(X)[:, 1]
df['ml_prediction'] = (df['ml_probability'] > 0.8).astype(int)

# –ö–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ–¥—Ö–æ–¥
df['final_prediction'] = ((df['ml_prediction'] == 1) | (df['rule_based_suspicious'] == 1)).astype(int)

print(f"‚úÖ –ú–æ–¥–µ–ª—å –æ–±—É—á–µ–Ω–∞ –∑–∞ {time.time()-start:.1f} —Å–µ–∫")
print(f"üìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
print(f"   ‚Ä¢ ML –ø–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö: {df['ml_prediction'].sum():,}")
print(f"   ‚Ä¢ –ü—Ä–∞–≤–∏–ª–∞ –ø–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö: {df['rule_based_suspicious'].sum():,}")
print(f"   ‚Ä¢ –ö–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö: {df['final_prediction'].sum():,}")
print(f"   ‚Ä¢ Precision ML: {precision_score(df['isFraud'], df['ml_prediction'], zero_division=0):.2%}")
print(f"   ‚Ä¢ Recall ML: {recall_score(df['isFraud'], df['ml_prediction'], zero_division=0):.2%}")
```

```python
print("=" * 60)
print("üìÅ –°–û–ó–î–ê–ù–ò–ï –§–ò–ù–ê–õ–¨–ù–û–ì–û CSV –§–ê–ô–õ–ê")
print("=" * 60)

# –§—É–Ω–∫—Ü–∏—è –æ–±—ä—è—Å–Ω–µ–Ω–∏–π –Ω–∞ –æ—Å–Ω–æ–≤–µ –Ω–æ–≤—ã—Ö –ø—Ä–∞–≤–∏–ª
def generate_explanation(row):
    explanations = []

    # –ü—Ä–∞–≤–∏–ª–æ 1: –ù—É–ª–µ–≤–æ–π –±–∞–ª–∞–Ω—Å + –∫—Ä—É–ø–Ω–∞—è —Å—É–º–º–∞ (>1 –º–ª–Ω)
    if row['rule1'] == 1:
        explanations.append(f"–ù—É–ª–µ–≤–æ–π –±–∞–ª–∞–Ω—Å + ${row['amount']:,.0f}")

    # –ü—Ä–∞–≤–∏–ª–æ 2: –°—É–º–º–∞ > 2 –º–ª–Ω
    if row['rule2'] == 1:
        explanations.append(f"–°—É–º–º–∞ > $2M (${row['amount']:,.0f})")

    # –ü—Ä–∞–≤–∏–ª–æ 3: –í—ã–≤–æ–¥ –≤—Å–µ—Ö –¥–µ–Ω–µ–≥ + —Å—É–º–º–∞ >500K
    if row['rule3'] == 1:
        explanations.append(f"–í—ã–≤–æ–¥ –≤—Å–µ—Ö –¥–µ–Ω–µ–≥ + >$500K")

    # –ü—Ä–∞–≤–∏–ª–æ 4: –ù–æ—á—å + –∫—Ä—É–ø–Ω–∞—è —Å—É–º–º–∞ (>1 –º–ª–Ω)
    if row['rule4'] == 1:
        explanations.append(f"–ù–æ—á—å + ${row['amount']:,.0f}")

    # –ü—Ä–∞–≤–∏–ª–æ 5: –ë—ã—Å—Ç—Ä—ã–µ –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–∏
    if row['rule5'] == 1:
        explanations.append("–ü–æ–≤—Ç–æ—Ä–Ω–∞—è —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏—è <1 —á–∞—Å–∞")

    # ML –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
    if row['ml_probability'] > 0.9:
        explanations.append(f"ML: {row['ml_probability']:.1%}")
    elif row['ml_probability'] > 0.8:
        explanations.append(f"ML: {row['ml_probability']:.1%}")

    # –ï—Å–ª–∏ –ø–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω–∞—è –Ω–æ –Ω–µ—Ç –ø—Ä–∞–≤–∏–ª
    if row['final_prediction'] == 1 and not explanations:
        explanations.append("–û–±–Ω–∞—Ä—É–∂–µ–Ω–æ —Å–∏—Å—Ç–µ–º–æ–π")

    return "; ".join(explanations) if explanations else "–ù–æ—Ä–º–∞–ª—å–Ω–∞—è —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏—è"

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–π DataFrame
final_results = pd.DataFrame({
    'transaction_id': range(1, len(df) + 1),
    'fraud_score': df['ml_probability'].round(4),
    'is_suspicious': df['final_prediction'],
    'explanation': df.apply(generate_explanation, axis=1)
})

# –î–æ–±–∞–≤–ª—è–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
for col in ['step', 'type', 'amount', 'nameOrig', 'nameDest', 'oldbalanceOrg', 'newbalanceOrig']:
    if col in df.columns:
        final_results[col] = df[col]

# –°–æ—Ö—Ä–∞–Ω—è–µ–º
csv_path = '/content/drive/MyDrive/fraud_detection_results.csv'
final_results.to_csv(csv_path, index=False)

print(f"‚úÖ CSV —Ñ–∞–π–ª —Å–æ–∑–¥–∞–Ω: {csv_path}")
print(f"üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:")
print(f"   ‚Ä¢ –í—Å–µ–≥–æ —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–π: {len(final_results):,}")
print(f"   ‚Ä¢ –ü–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö: {final_results['is_suspicious'].sum():,}")
print(f"   ‚Ä¢ –î–æ–ª—è: {final_results['is_suspicious'].mean()*100:.2f}%")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø—Ä–∏–º–µ—Ä—ã
print(f"\nüîç –ü—Ä–∏–º–µ—Ä—ã –∑–∞–ø–∏—Å–µ–π:")
suspicious_examples = final_results[final_results['is_suspicious'] == 1].head(3)
for idx, row in suspicious_examples.iterrows():
    print(f"   üö® ID {row['transaction_id']}: {row['type']} ${row['amount']:,.0f}")
    print(f"      Fraud Score: {row['fraud_score']:.4f}")
    print(f"      –û–±—ä—è—Å–Ω–µ–Ω–∏–µ: {row['explanation']}")
    print()
```

```python
%%writefile streamlit_app.py

# streamlit_app.py ‚Äî –ú–∏–Ω–∏-–∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å FraudDetect —Å –ø–æ–ª–Ω—ã–º–∏ –ø—Ä–∞–≤–∏–ª–∞–º–∏
import streamlit as st
import pandas as pd
import os
from datetime import datetime
import io
import gzip

st.set_page_config(page_title="FraudDetect Demo", layout="wide")
st.title("üîç FraudDetect ‚Äî –ú–∏–Ω–∏ –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å —Å ML + –ø—Ä–∞–≤–∏–ª–∞–º–∏")

# ---------------------- –ü—É—Ç—å –∫ CSV ----------------------
DRIVE_CSV_PATH = "/content/drive/MyDrive/fraud_detection_results.csv"

if not os.path.exists(DRIVE_CSV_PATH):
    st.error(f"CSV —Ñ–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω –Ω–∞ Drive: {DRIVE_CSV_PATH}")
    st.stop()

# –ó–∞–≥—Ä—É–∂–∞–µ–º CSV
df = pd.read_csv(DRIVE_CSV_PATH)
st.success(f"CSV –∑–∞–≥—Ä—É–∂–µ–Ω —Å Drive: {len(df):,} —Å—Ç—Ä–æ–∫")
st.dataframe(df.head(5))

# ---------------------- –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –æ–±—ä—è—Å–Ω–µ–Ω–∏–π ----------------------
def generate_explanation(row):
    explanations = []

    # ML –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
    if row['is_suspicious'] == 1:
        if row['fraud_score'] > 0.9:
            explanations.append("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ñ—Ä–æ–¥–∞ >90%")
        elif row['fraud_score'] > 0.8:
            explanations.append("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ñ—Ä–æ–¥–∞ >80%")
        elif row['fraud_score'] > 0.7:
            explanations.append("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ñ—Ä–æ–¥–∞ >70%")

    # –°—É–ø–µ—Ä-—Å—Ç—Ä–æ–≥–∏–µ –ø—Ä–∞–≤–∏–ª–∞
    for i in range(1,6):
        rule = f'r{i}'
        if rule in row and row[rule] == 1:
            if rule == 'r1':
                explanations.append("TRANSFER + –æ—á–µ–Ω—å –∫—Ä—É–ø–Ω–∞—è —Å—É–º–º–∞ (>500K)")
            elif rule == 'r2':
                explanations.append("–°—É–º–º–∞ > 2 –º–ª–Ω")
            elif rule == 'r3':
                explanations.append("–ù—É–ª–µ–≤–æ–π –±–∞–ª–∞–Ω—Å + —Å—É–º–º–∞ >500K")
            elif rule == 'r4':
                explanations.append("–í—ã–≤–æ–¥ –≤—Å–µ—Ö –¥–µ–Ω–µ–≥ (100%) + —Å—É–º–º–∞ >100K")
            elif rule == 'r5':
                explanations.append("–ù–æ—á—å + —Å—É–º–º–∞ >500K")

    return "; ".join(explanations) if explanations else "–ù–æ—Ä–º–∞–ª—å–Ω–∞—è —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏—è"

df['explanation'] = df.apply(generate_explanation, axis=1)

# ---------------------- –í—ã–≤–æ–¥ —Ç–æ–ø —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–π ----------------------
display_cols = [
    'step', 'type', 'amount', 'nameOrig', 'oldbalanceOrg', 'newbalanceOrig',
    'nameDest', 'oldbalanceDest', 'newbalanceDest', 'isFraud',
    'fraud_score', 'is_suspicious', 'explanation'
]
display_cols = [c for c in display_cols if c in df.columns]

st.write("### üîé –¢–æ–ø 20 –ø–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–π")
st.dataframe(df[df['is_suspicious']==1][display_cols].head(20))

st.write("### üîé –¢–æ–ø 20 –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏–π")
st.dataframe(df[df['is_suspicious']==0][display_cols].head(20))

# ---------------------- –°–∫–∞—á–∞—Ç—å CSV —Å gzip ----------------------
buf = io.BytesIO()
with gzip.GzipFile(fileobj=buf, mode='wb') as f:
    f.write(df[display_cols].to_csv(index=False).encode('utf-8'))

st.download_button(
    "üì• –°–∫–∞—á–∞—Ç—å CSV (gzip)",
    data=buf.getvalue(),
    file_name=f"fraud_results_full_{datetime.now().strftime('%Y%m%d_%H%M')}.csv.gz"
)

st.markdown("---")
st.markdown("‚ö° –í —Ç–∞–±–ª–∏—Ü–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã –∫–ª—é—á–µ–≤—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
st.markdown("- `fraud_score` ‚Äî –æ—Ü–µ–Ω–∫–∞ ML –º–æ–¥–µ–ª–∏")
st.markdown("- `is_suspicious` ‚Äî –ø–æ–¥–æ–∑—Ä–∏—Ç–µ–ª—å–Ω–∞—è —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏—è")
st.markdown("- `explanation` ‚Äî –ø–æ—á–µ–º—É —Ç—Ä–∞–Ω–∑–∞–∫—Ü–∏—è –æ—Ç–º–µ—á–µ–Ω–∞")
st.markdown("- –û—Å—Ç–∞–ª—å–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏ ‚Äî –∫–æ–Ω—Ç–µ–∫—Å—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ (step, type, amount, nameOrig, nameDest, –±–∞–ª–∞–Ω—Å—ã –∏ —Ç.–¥.)")
```
